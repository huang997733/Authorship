```python
import json
import numpy as np
```

## Data preprocessing

```python
with open("train.json","r") as f:
    train = json.load(f)
```

```python
with open("test.json","r") as f:
    test = json.load(f)
```

### process train data

```python
y = []
for data in train:
    authors = np.array(data['authors'])
    prolific_authors = authors[authors<100]
    label = np.zeros(101)
    if len(prolific_authors) == 0:
        label[-1] = 1.
    else:
        label[prolific_authors] = 1.
    y.append(label)
y = np.array(y)
```

```python
from gensim.models import Word2Vec
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
```

```python
words = []
for data in train:
    title = data['title']
    abstract = data['abstract']
    words.append(title+abstract)
```

```python
documents = [TaggedDocument(doc, [i]) for i, doc in enumerate(words)]
d2v = Doc2Vec(documents, vector_size=300, window=2, min_count=1)
```

```python
doc = []
for data in train:
    title = data['title']
    abstract = data['abstract']
    total = title + abstract
    vec = d2v.infer_vector(np.asarray(total,dtype='str'))
    doc.append(vec)
doc = np.array(doc)
```

```python
venue = []
for data in train:
    v = data['venue']
    if v == "":
        venue.append([465])
    else:
        venue.append([v])
venue = np.array(venue)
```

```python
cas = {}
for data in train:
    a = np.array(data['authors'])
    pa = a[a<100]
    ca = a[a>=100]
    if len(pa) != 0:
        for author in ca:
            cas[author] = 1
```

```python
ca_list = list(cas.keys())
```

```python
len(ca_list)
```

```python
ca_index = {}
for i, d in enumerate(ca_list):
    ca_index[d] = i
```

```python
coauthors = []
for data in train:
    a = np.array(data['authors'])
    pa = a[a<100]
    ca = a[a>=100]
    temp = np.zeros(6576)
    if len(pa) == 0:
        temp[-1] = 1.
    else:
        for x in ca:
            temp[ca_index[x]] = 1.
    coauthors.append(temp)
coauthors = np.array(coauthors)
```

```python
x = np.concatenate((doc,venue), axis=1)
```

```python
x = np.concatenate((x,coauthors), axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

### process test data

```python
doc = []
for data in test:
    title = data['title']
    abstract = data['abstract']
    total = title + abstract
    vec = d2v.infer_vector(np.asarray(total,dtype='str'))
    doc.append(vec)
doc = np.array(doc)
```

```python
venue = []
for data in test:
    v = data['venue']
    if v == "":
        venue.append([465])
    else:
        venue.append([v])
venue = np.array(venue)
```

```python
coauthors = []
for data in test:
    a = np.array(data['coauthors'])
    temp = np.zeros(6576)
    for i in a:
        if i in ca_index: 
            temp[ca_index[i]] = 1.
    coauthors.append(temp)
coauthors = np.array(coauthors)
```

```python
test = np.concatenate((doc,venue), axis=1)
test = np.concatenate((test,coauthors), axis=1)
```

```python
scaler = StandardScaler()
test = scaler.fit_transform(test)
```

## Train models

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import f1_score
import pandas as pd
```

```python
x.shape
```

```python
y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 90051)
```

```python
weights = {}
for i in range(101):
    weights[i] = 10.
weights[100] = 2.5
```

```python
model = Sequential()
model.add(Dense(2048,input_dim=x.shape[1],activation='tanh'))
model.add(Dropout(0.3))
model.add(Dense(512,activation='tanh'))
model.add(Dropout(0.3))
model.add(Dense(512,activation='tanh'))
model.add(Dropout(0.3))
model.add(Dense(256,activation='tanh'))
model.add(Dropout(0.3))
model.add(Dense(y.shape[1],activation='sigmoid'))

model.compile(loss = 'binary_crossentropy', optimizer = Adam())
model.fit(x_train, y_train, epochs=100, batch_size=128, class_weight=weights)
```

```python
train_pred = model.predict(x_train)
train_pred = np.where(train_pred>0.5,1,0)
```

```python
f1_score(y_true=y_train, y_pred=train_pred,average='samples')
```

```python
test_pred = model.predict(x_test)
test_pred = np.where(test_pred>0.5,1,0)
```

```python
f1_score(y_true=y_test, y_pred=test_pred,average='samples')
```

```python
result = model.predict(test)
```

```python
idx = [np.where(p>0.5) for p in result]
df = pd.DataFrame(idx, columns=['Predict'])
df
```

```python
df.to_csv('result.csv')
```

### prolific and coauthor dictionary

```python
authors_dict = {}
for data in train:
    authors = np.array(data['authors'])
    year = np.array(data['year'])
    prolific_authors = authors[authors<100]
    co_authors = authors[authors>=100]
    if year > 16:
        
        for a in co_authors:
            if a in authors_dict and prolific_authors.size != 0:
                n = 0
                for pa in prolific_authors:
                    if n<3: 
                        n += 1
                        np.append(authors_dict[a],pa) if pa not in authors_dict[a] else authors_dict[a]
            elif a not in authors_dict and prolific_authors.size != 0:
                authors_dict[a] = prolific_authors
```

```python
authors_dict
```

```python
result = []
for data in test:
    cas = np.array(data['coauthors'])
    r = []
    for ca in cas:
        if ca in authors_dict:
            for a in authors_dict[ca]:
                r.append(a) if a not in r else r
    result.append(np.array(r))
    
```

```python
result
```

```python
import random

final = [] 
for i, d in enumerate(idx):
    if d[0].size == 0 or d[0][0] == 100 and random.uniform(0, 1) > 0.7:
        temp = np.array([])
        if result[i].size != 0:
            for a in result[i]:
                np.append(temp,a)
        final.append((np.append(d[0],a),))
    else:
        final.append(d)
                
```
